```python
import pandas as pd
import numpy as np
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

```python
file_path = 'D:/Study/17-Fall/EE-628/Final Project/'
train_file_name = 'train.json'
test_file_name = 'test.json'

train = pd.read_json(train_file_name)
test = pd.read_json(test_file_name)
```

```python
data = [go.Bar(
            x = ['ship', 'iceberg'],
            y = train['is_iceberg'].value_counts().values,
            text='Distribution of target variable'
    )]

layout = go.Layout(
    title='Target variable distribution'
)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='basic-bar')
```

```python
plt.figure(figsize=(16,8))
train_dropna = train[(True ^ train['inc_angle'].isin(['na']))]
train_dropna['inc_angle'].plot(kind='kde')
test['inc_angle'].plot(kind='kde')
plt.legend(('train', 'test'),loc='best',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16,8))
train_dropna.inc_angle[train_dropna.is_iceberg == 0].plot(kind='kde')
train_dropna.inc_angle[train_dropna.is_iceberg == 1].plot(kind='kde')
plt.legend(('ship', 'iceberg'),loc='best',fontsize=20)
plt.show()
```

```python
print ('Range of inc_angle in train_data: (%.2f,%.2f)' %(train_dropna.inc_angle.min(),train_dropna.inc_angle.max())) 
print ('Range of inc_angle in test_data: (%.2f,%.2f)' %(test.inc_angle.min(),test.inc_angle.max()))
```

```python
X_band_1=np.array([np.array(band).astype(np.float32).reshape(75, 75) for band in train["band_1"]])
X_band_2=np.array([np.array(band).astype(np.float32).reshape(75, 75) for band in train["band_2"]])
```

```python
def plotmy3d(c, name):
    data = [
        go.Surface(
            z=c
        )
    ]
    layout = go.Layout(
        title=name,
        autosize=False,
        width=700,
        height=700,
        margin=dict(
            l=65,
            r=50,
            b=65,
            t=90
        )
    )
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig)
```

```python
plotmy3d(X_band_1[11,:,:], 'ship')
```

```python
plotmy3d(X_band_1[12,:,:], 'iceberg')
```

```python
# ship
fig = plt.figure(figsize=(16,8))  
ax1 = fig.add_subplot(1,2,1)  
ax1.imshow(X_band_1[11,:,:], cmap=plt.cm.gray)  
ax2 = fig.add_subplot(1,2,2)  
ax2.imshow(X_band_2[11,:,:], cmap=plt.cm.gray)  
plt.show()
```
